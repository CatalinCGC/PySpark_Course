```python
emp_data_1.printSchema()
emp_data_2.printSchema()
```

```python
# Union and Union all

emp = emp_data_1.unionAll(emp_data_2)

emp.show()
```

```python
# sort data

from pyspark.sql.functions import desc, asc

emp_sorted = emp.sort(asc("salary"))

emp_sorted.show()
```

```python
# Aggregations

from pyspark.sql.functions import count

emp_count = emp.groupBy("department_id").agg(count("employee_id").alias("total_dept_count"))

emp_count.show()
```

```python
from pyspark.sql.functions import sum

emp_sum = emp.groupBy("department_id").agg(sum("salary").alias("total_dept_salary"))

emp_sum.show()
```

```python
# Having clause

from pyspark.sql.functions import avg

emp_avg = emp.groupBy("department_id").agg(avg("salary").alias("avg_dept_salary")).where("avg_dept_salary > 50000")

emp_avg.show()
```
