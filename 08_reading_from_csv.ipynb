```python
df = spark.read.format("csv").option("header", True).option("inferSchema", True).load("/Volumes/workspace/pyspark/input/emp.csv")
```

```python
df.show()
```

```python
df.printSchema()
```

```python
# Reading with Schema
_schema = "employee_id int, department_id int, name string, age int, gender string, salary double, hire_date date"

df_schema = spark.read.format("csv").option("header",True).schema(_schema).load("/Volumes/workspace/pyspark/input/emp.csv")
```

```python
df_schema.printSchema()
```

```python
df_schema.show()
```

```python

# Handle BAD records - PERMISSIVE (Default mode)

_schema = "employee_id int, department_id int, name string, age int, gender string, salary double, hire_date date, bad_record string"

df_p = spark.read.format("csv").schema(_schema).option("columnNameOfCorruptRecord", "bad_record").option("header", True).load("/Volumes/workspace/pyspark/input/emp_new.csv")
```

```python
df_p.printSchema()
```

```python
df_p.show(truncate=False)
```

```python
df_p.where("bad_record is null").show(truncate=False)
```

```python
# Handle BAD records - DROPMALFORMED
_schema = "employee_id int, department_id int, name string, age int, gender string, salary double, hire_date date"

df_m = spark.read.format("csv").option("header", True).option("mode", "DROPMALFORMED").schema(_schema).load("/Volumes/workspace/pyspark/input/emp_new.csv")

df_m.show()
```

```python
# Handle BAD records - FAILFAST
_schema = "employee_id int, department_id int, name string, age int, gender string, salary double, hire_date date"

df_m = spark.read.format("csv").option("header", True).option("mode", "FAILFAST").schema(_schema).load("/Volumes/workspace/pyspark/input/emp_new.csv")

df_m.show()
```

```python
# Multiple options

_options = {
    "header" : "true",
    "inferSchema" : "true",
    "mode" : "PERMISSIVE"
}

df = (spark.read.format("csv").options(**_options).load("/Volumes/workspace/pyspark/input/emp_new.csv"))

df.show()
```
